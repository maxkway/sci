```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
np.random.rand?
```

```python
N=20
sigma = 10
```

```python
lambda_star = np.random.choice(np.array([-1,1]))
lambda_star
```

```python
sample_x = sigma*np.random.normal(size=N) + lambda_star
```

```python
plt.figure()
plt.grid()
plt.plot(sample_x,'.')
ax = plt.gca()
plt.xlabel(r'$i$')
plt.ylabel(r'$x_i/\bar{x}$')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
```

```python
plt.hist(sample_x, 4)
```

```python
sample_x.mean()
```

```python
def get_P_x_A(x):
    return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-1/2*(x-1)**2/sigma**2)
def get_P_x_B(x):
    return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-1/2*(x+1)**2/sigma**2)
```

```python
p_A = 1/2
p_B = 1/2
plt.figure()
n=0
plt.scatter(n, p_A, color='red')
for x in sample_x:
    p_A = get_P_x_A(x) * p_A / ( get_P_x_A(x) * p_A + get_P_x_B(x) * p_B)
    p_B = 1 - p_A
    n +=1
    plt.scatter(n, p_A, color='red')
ax = plt.gca()
plt.xlabel(r'$n$')
plt.ylabel(r'$\tilde{P}_A$')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
```
