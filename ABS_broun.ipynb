```python
import numpy as np
```

$$
dx = - sgn(x)u_0dt + \sqrt{2dt}d\varepsilon
$$

```python
dt = 1e-3
T_max = 1e3
u0 = 1e-1
plt.figure()
for m in range(10):
    t = 0
    x=0
    x_array = [x]
    t_array =[t]
    while x<1 and t<T_max:
        x += -np.sign(x)*u0*dt + np.sqrt(2*dt)*np.random.normal()
        t += dt
        x_array.append(x)
        t_array.append(t)
    x_array = np.array(x_array)
    t_array = np.array(t_array)
    plt.plot(t_array[::10], x_array[::10])
```

```python
dt = 1e-3
T_max = 1e3
u0 = 1e-1
T_array = []
for m in range(1000):
    t = 0
    x=0
    x_array = [x]
    t_array =[t]
    while x<1 and t<T_max:
        x += -np.sign(x)*u0*dt + np.sqrt(2*dt)*np.random.normal()
        t += dt
    T_array.append(t)
```

```python
T_array = np.array(T_array)
plt.hist(T_array, 100)
plt.yscale('log')
plt.title(fr'$<T> = {np.array(T_array).mean():.2f}, T_m = {np.median(T_array):.2f}$')#, density=True)
```

```python
plt.hist(T_array[T_array < 50], 20)
plt.yscale('log')
np.array(T_array).mean()#, density=True)
```

И число шагов с тем же шагом

```python
dt = 1e-3
T_max = 1e4
u0 = 1e-1
T_array = []
for m in range(50000):
    t = 0
    x=0
    x_array = [x]
    t_array =[t]
    while x<1 and t<T_max:
        x += -np.sign(x)*u0*dt + np.sqrt(2*dt)*np.random.normal()
        t += dt
    T_array.append(t)
```

```python
T_array = np.array(T_array)
plt.hist(T_array, 100)
plt.yscale('log')
plt.title(fr'$<T> = {np.array(T_array).mean():.2f}, T_m = {np.median(T_array):.2f}$')#, density=True)
```

посмотрим вблизи нуля:

```python
plt.hist(T_array[T_array < 3], 100)
t_ = np.linspace(0,3, 500)
plt.plot(t_, 1.5e3*np.exp(-1/4/t_)/t_**(3/2)/2/np.sqrt(np.pi), label='levi')
plt.legend()
```
